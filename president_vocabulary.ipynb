## Import Libraries

```python
from nltk.corpus import inaugural
from functions.president_helper import process_speeches, merge_speeches, get_president_sentences, get_presidents_sentences, most_frequent_words, create_wordcloud, map_president_name
import gensim.models
```

## All Presidents

### Most frequently used words

```python
# Get the file IDs for the inaugural speeches
file_ids = inaugural.fileids()

# Create a list to store the speeches
speeches = []

# Read the content of each file and add it to the speeches list
for file_id in file_ids:
    speech = inaugural.raw(file_id)
    speeches.append(speech)

# Process the speeches into a list of sentences and words
processed_speeches = process_speeches(speeches)

# Merge the speeches into a single list of sentences and words
all_sentences = merge_speeches(processed_speeches)

# Get the most frequent words
most_freq_words = most_frequent_words(all_sentences)
print("Most frequent words:", most_freq_words)
create_wordcloud(most_freq_words)
```

**Summary:**  
1. **Government and People**: The words "government" and "people" are among the most frequently used, indicating a strong emphasis on the relationship between the government and the citizens. It reflects the core democratic principles of representation and governance.

3. **Freedom and Liberty**: The presence of "freedom" and "liberty" reflects a commitment to individual rights and civil liberties, which are foundational to the American identity.

6. **Progress and Hope**: The words "progress" and "hope" indicate a forward-looking perspective, emphasizing the pursuit of progress, prosperity, and a better future for the country.

7. **Duty and Responsibility**: The terms "duty" and "responsibility" reflect the president's commitment to their role as a leader and the responsibilities that come with it.

8. **Unity and Common Good**: Phrases like "together" and "common" emphasize the importance of unity and working collectively for the common good of the nation.

### Words most similar to "freedom" 

```python
# Create a Word2Vec model for all presidents
all_prez_embeddings = gensim.models.Word2Vec(all_sentences, vector_size=96, window=5, min_count=1, workers=2, sg=1)
all_prez_embeddings.build_vocab(all_sentences)
all_prez_embeddings.train(all_sentences, total_examples=all_prez_embeddings.corpus_count, epochs=all_prez_embeddings.epochs)

# Find words similar to "freedom" for all presidents
similar_to_freedom = all_prez_embeddings.wv.most_similar("freedom", topn=20)
print("Words similar to 'freedom' for all presidents:", similar_to_freedom)
create_wordcloud(similar_to_freedom)
```
